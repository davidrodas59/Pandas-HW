```python
#Purchasing Analysis (Age)
```

```python
#bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

bin_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_data["bin"] = pd.cut(purchase_data["Age"], bins, labels=bin_names)
purchase_data.head(10)

```

```python
#group by age
#Bin the purchase_data data frame by age
groupby_age =age_data.groupby('bin')
groupby_age.head()

#purchase by group
purchase_by_age = groupby_age['Purchase ID'].count()
#avg purchase price
avg_purchase_price = groupby_age['Price'].mean()

 #total per person

total_purchase = groupby_age['Price'].sum()

#avg total per person in group

avg_total = total_purchase/avg_purchase_price

Purchase_Analysis = pd.DataFrame({'Purchase Count': purchase_by_age,
                                  'Average Purchase Price': avg_purchase_price,
                                 'Total Purchase Value': total_purchase,
                                 'Avg Total Purchase per Person': avg_total})
Purchase_Analysis.head(10)

#'Purchase Count': purchase_by_age, , index=[0]
                                 
                                 
```

```python
# Top Spenders
```

```python

purchase = purchase_data.groupby(['SN'])
purchase.head(5)
#purchases made
purchase_count = purchase['Purchase ID'].count()
#avg price
avg_purchase_price = purchase['Price'].mean()
#total purchase value
total_purchase = purchase['Price'].sum()
top_spender = pd.DataFrame({"Purchase Count":purchase_count,
                            'Average Purchase Price':avg_purchase_price,
                            'Total Purchase Value': total_purchase
    
})
top_spender
#sort top spender
sorted_spenders = top_spender.sort_values(["Total Purchase Value"], ascending=False).head()
sorted_spenders
```

```python
#Most Popular Items
```

```python
# Retrieve the Item ID, Item Name, and Item Price columns

# Group by Item ID and Item Name. 
#Perform calculations to obtain purchase count, item price, and total purchase value

id_name_price = purchase_data[['Item ID', 'Item Name', 'Price']]
gby_id_name = id_name_price.groupby(['Item ID', 'Item Name'])

count = gby_id_name['Price'].count()


purchase_value = (gby_id_name['Price'].sum())
purchase_value
price_per_item = purchase_value/count

most_popular = pd.DataFrame({ 'Purchase Count': count,
                             'Price': price_per_item,
                             'Total Purchase Value': purchase_value
    
})
most_popular.sort_values(by='Total Purchase Value', ascending= False).head()
```
